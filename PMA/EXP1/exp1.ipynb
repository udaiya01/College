```python
import pandas as pd
import numpy as np

# Creating the dataset directly in the code
data = {'X': [1, 2, 3, 4, 5],
        'Y': [2.5, 5, 7.5, 10, 12.5]}

# Creating a pandas dataframe
df = pd.DataFrame(data)

# Calculate statistical measures
min_values = df.min()
max_values = df.max()
mean_values = df.mean()
median_values = df.median()
quantiles = df.quantile([0.25, 0.5, 0.75])
std_values = df.std()
var_values = df.var()
summary = df.describe()

# Print results

print("Min values:\n", min_values)
print("\nMax values:\n", max_values)
print("\nMean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nQuantiles:\n", quantiles)
print("\nStandard Deviation values:\n", std_values)
print("\nVariance values:\n", var_values)
print("\nSummary:\n", summary)
```
