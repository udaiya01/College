```python
# Step 1: Install PySpark
# pip install pyspark

# Step 2: Import Libraries
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
import pandas as pd

# Step 3: Set up Spark Session
spark = SparkSession.builder.appName("IrisDecisionTree").getOrCreate()

# Step 4: Load Iris Data
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Convert to Spark DataFrame
df = spark.createDataFrame(iris_df)

# Step 5: Prepare Features and Target
assembler = VectorAssembler(inputCols=iris.feature_names, outputCol="features")
df = assembler.transform(df)
df = df.withColumnRenamed("target", "label")

# Step 6: Split Data
train, test = df.randomSplit([0.8, 0.2], seed=42)

# Step 7: Train the Decision Tree Model
dt = DecisionTreeClassifier(featuresCol="features", labelCol="label")
model = dt.fit(train)

# Step 8: Make Predictions
predictions = model.transform(test)

# Step 9: Evaluate Model
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)

# Optional: Print Accuracy
print(f"Test Accuracy: {accuracy}")

# Step 10: Stop Spark Session
spark.stop()
```
