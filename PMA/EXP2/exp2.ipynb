```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample dataset
data = {
    'Temperature': [30, 32, 35, 36, 34, 30, 28, 25, 26, 22],
    'Humidity': [45, 50, 55, 40, 42, 38, 37, 35, 33, 31],
    'City': ['City A', 'City B', 'City C', 'City D', 'City E',
             'City F', 'City G', 'City H', 'City I', 'City J'],
    'Rainfall': [100, 150, 80, 90, 110, 95, 75, 60, 70, 50],
    'Date': pd.date_range(start='2024-08-01', periods=10, freq='D')
}
df = pd.DataFrame(data)

# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature', y='Humidity', data=df, s=100)
plt.title('Temperature vs. Humidity ')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
sns.despine()
plt.show()

# Horizontal Bar Chart without hue
plt.figure(figsize=(8, 6))
sns.barplot(x='Rainfall', y='City', data=df, orient='h')
plt.title('Rainfall in Different Cities ')
plt.xlabel('Rainfall (mm)')
plt.ylabel('City')
sns.despine()
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Temperature'], bins=5, kde=True)
plt.title('Distribution of Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
sns.despine()
plt.show()

# Line Graph (Time Series Data Visualization)
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Temperature'], marker='o', color='b', linestyle='-')
plt.title('Temperature Over Time ')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
sns.despine()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample dataset
data = {
    'Temperature': [30, 32, 35, 36, 34, 30, 28, 25, 26, 22],
    'Humidity': [45, 50, 55, 40, 42, 38, 37, 35, 33, 31],
    'City': ['City A', 'City B', 'City C', 'City D', 'City E',
             'City F', 'City G', 'City H', 'City I', 'City J'],
    'Rainfall': [100, 150, 80, 90, 110, 95, 75, 60, 70, 50],
    'Date': pd.date_range(start='2024-08-01', periods=10, freq='D')
}
df = pd.DataFrame(data)

# Scatter Plot
sns.scatterplot(x='Temperature', y='Humidity', data=df)
plt.title('Temperature vs. Humidity')
plt.show()

# Horizontal Bar Chart
sns.barplot(x='Rainfall', y='City', data=df, orient='h')
plt.title('Rainfall in Different Cities')
plt.show()

# Histogram
sns.histplot(df['Temperature'], bins=5)
plt.title('Distribution of Temperature')
plt.show()

# Line Graph
plt.plot(df['Date'], df['Temperature'], marker='o')
plt.title('Temperature Over Time')
plt.xticks(rotation=45)
plt.show()
```
