```python
import numpy as np
from scipy.stats import chi2_contingency, f_oneway

# ----- Chi-Square Test -----
# Sample Data for Chi-Square Test (Contingency Table)
observed = np.array([[30, 20],
                     [20, 25],
                     [50, 35]])

# Perform the Chi-Square Test
chi2, p_chi2, dof, expected = chi2_contingency(observed)

print("----- Chi-Square Test Results -----")
print(f"Chi-Square Statistic: {chi2}")
print(f"p-value: {p_chi2}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:{expected}")
print(expected)
print("\n")

# ----- ANOVA F-Test -----
# Sample Data for ANOVA F-Test
class1 = [85, 88, 90, 87]
class2 = [78, 82, 79, 85]
class3 = [92, 94, 89, 91]

# Perform the ANOVA F-Test
f_statistic, p_anova = f_oneway(class1, class2, class3)

print("----- ANOVA F-Test Results -----")
print(f"F-Statistic: {f_statistic}")
print(f"p-value: {p_anova}")
```
