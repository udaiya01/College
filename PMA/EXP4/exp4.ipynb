```python
from sklearn.decomposition import PCA
import numpy as np

# Example dataset
X = np.array([[2.5, 2.4, 3.5],
              [0.5, 0.7, 0.6],
              [2.2, 2.9, 3.1],
              [1.9, 2.2, 2.7],
              [3.1, 3.0, 3.4]])

# Initialize PCA with the number of components
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print(X_reduced)
```
