```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

data = {
'Age': [25, 45, 35, 50, 22, 30, 28, 60, 40, 55],
'Income': ['Low', 'Medium', 'High', 'Medium', 'Low', 'High', 'Low', 'High', 'Medium', 'High'],
'Education': ['High', 'Medium', 'Medium', 'High', 'High', 'Medium', 'Low', 'High', 'Low', 'High'],
'Buy_Product': ['No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

df = pd.get_dummies(df, columns=['Income', 'Education'], drop_first=True)

X = df.drop('Buy_Product', axis=1)

y = df['Buy_Product'].apply(lambda x: 1 if x == 'Yes' else 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the Decision Tree: {accuracy:.2f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# URL of the published CSV file
url = 'https://example.com/path-to-your-csv-file.csv'

# Read the data from the URL
df = pd.read_csv(url)

# One-hot encode categorical features
df = pd.get_dummies(df, columns=['Income', 'Education'], drop_first=True)

# Separate features and target variable
X = df.drop('Buy_Product', axis=1)
y = df['Buy_Product'].apply(lambda x: 1 if x == 'Yes' else 0)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the Decision Tree model
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# Make predictions and calculate accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy of the Decision Tree: {accuracy:.2f}")
```
